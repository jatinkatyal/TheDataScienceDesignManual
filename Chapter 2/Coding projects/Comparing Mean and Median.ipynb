```python
import pandas as pd
```

```python
df = pd.read_csv('../../Chapter 1/Best Seller Analysis/datasheet.csv',header=None, names=['BookTitle','Author','Category','Price','CaptureTimestamp','Rank'])
df.head(5)
```

```python
mean = df.groupby('Category').Price.mean()
median = df.groupby('Category').Price.median()
meanVsMedian = pd.merge(mean,median,on='Category')
meanVsMedian=meanVsMedian.rename(columns={'Price_x':'MeanPrice','Price_y':'MedianPrice'})
meanVsMedian

```
