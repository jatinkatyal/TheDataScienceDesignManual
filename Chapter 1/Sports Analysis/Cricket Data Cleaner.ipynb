I got a file a data set of all cricket players which was scrapped from cricinfo on kaggle. I don't want to use all of it and will focus on only Indian players who played international matches for India A.

```python
# Importing libraries i'll require
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cd = pd.read_csv('cricket_data.csv')
cd.shape
```

```python
cd.head(5)
```

That's a good sized dataset, let's see what all columns we have.

```python
#pd.set_option('display.max_rows',181) #Setting display size as output is more than default 10
cd.dtypes
```

I don't need all those columns, let's take only what we think is needed.

```python
cd = cd[['NAME','COUNTRY','Major teams','Playing role',
    'BATTING_T20Is_Inns','BATTING_T20Is_Runs','BATTING_T20Is_Ave','BATTING_T20Is_100','BATTING_T20Is_50',
    'BOWLING_T20Is_Inns','BOWLING_T20Is_Ave','BOWLING_T20Is_Econ','BOWLING_T20Is_SR']]
```

I am looking for only Indian players so let's remove other nationality players first.

```python
cd = cd[cd['COUNTRY']=='India']
cd.shape
```

We already reduced from 90K to 13K, lets look at only international players.

```python
cd[cd['Major teams']=='India A']
```

Odd i was expecting results, so the team column has a list of teams a player has played for and it can't just India A. Can't filter it directly.

```python
cd = cd[cd['Major teams'].isna()==False] #removing players with no teams
cd = cd[cd['Major teams'].str.contains('India A') & cd['Major teams'].str.contains('Women')==False]
```

Many indian players we selected haven't played a T20 international. Let's remove them.

```python
indianT20Players=cd[ (cd['BOWLING_T20Is_Inns'].isna()==False) | (cd['BATTING_T20Is_Inns'].isna()==False) ]
indianT20Players.to_csv('IndianT20IPlayers.csv')
indianT20Players.shape
```
